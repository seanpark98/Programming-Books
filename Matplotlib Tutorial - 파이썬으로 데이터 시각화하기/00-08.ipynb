```python
import matplotlib.pyplot as plt
import numpy as np
```

### 00-01

```python
"""
Without x values, plot() assumes the inputs as y values and 
automatically generates x values incrementally starting from 0 
"""
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
#first list is x, second list is y
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
"""
3rd argument in plot() determines the market.
- First character is the color and the second is shape of the marker.
ex. bo, r--, g^, 
*default is b-
"""
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

"""
Every sequence gets automatically converted into numpy arrays.
You can plot multiple plots using plot(x1,y1, style1, x2, y2, style2)
"""
plt.plot(t, t, 'ro-', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 02

```python
import matplotlib.pyplot as plt


"""You can pass in indexable objects such as dictionaries in plot() too"""
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

### 03. axis

```python
"""
first argument of label sets the name of axis.
labelpad sets padding of axis.
fontdict sets the font and color, and loc sets the location.
"""

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict={'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14}, loc='right')
plt.ylabel('Y-Axis', labelpad=20, fontdict={'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 'xx-large'}, loc = 'top')
plt.show()
```

### 04. legend

```python
"""
legends can be set by label. The location of the legend can also be specified
"""
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#plt.legend(loc=(0.6, 0.6))
plt.legend(loc =('upper left'))

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
"""
loc = 'best' places the legend at the best location
ncol is the number of columns and you can also set the fontsize
frameon sets the borders. There's also face parameters such as 
facecolor, edgecolor, borderpad, labelspacing.
"""
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)

plt.show()
```

### 05. axis range

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5]) #([min, max])     
plt.ylim((0, 20)) # plt.ylim(0,20)
#or you can use plt.axis([0,5,0,20]) [xmin, xmax, ymin, ymax]
plt.show()
```

### 06. linestyle

```python
"""shape of line can be determined as the 3rd argument and 
color can also be specified
or you can use linestyle = to determine the shape 
"""

plt.plot([1, 2, 3], [4, 4, 4], 'b-', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
#line style can also be set by tuples (0, (width, gap)) or (0, (width,gap, width,gap))

plt.plot([1, 2, 3], [4, 4, 4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))')
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
#you can also set the shape of capstyle with teh solid_capstyle argument

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

### 07. marker

```python
"""linestyles can also be set with the marker """
#plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo', linestyle='dashed')   
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo--')    
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#you can also use the marker parameter
plt.plot([4, 5, 6], marker="h")
plt.plot([3, 4, 5], marker="d")
plt.plot([2, 3, 4], marker="x")
plt.plot([1, 2, 3], marker=11)
plt.plot([0, 1, 2], marker='$Z$')
plt.show()
```

### 08. linecolor

```python
"""
linecolors can be set as 3rd argument (color, marker, linestyle) or with the color argument.
basic colors: b,g,r; Tableau colors: tab:pink, tab:brown; and CSS colors: bisque, seagreen can be used.
"""

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], 'b')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], color='limegreen')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], color='#e35f62',marker='o', linestyle='--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
